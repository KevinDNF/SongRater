```python
#--------Account-Details--------
username="21fm7pkvrha54o2nt7kiyqema"
cID = "020a9594aba74d13bf3312c48d9fea44"
secret = "f923ed27f36048dab8229e13ac999f1c"
scope = "user-top-read"
#scope = "user-follow-read user-library-read user-top-read user-read-recently-played"
```

```python
#--------Getting-Token---------
try:                                                                            
    token = util.prompt_for_user_token(username,scope,client_id=cID,client_secret=secret ,redirect_uri="http://localhost/")
except:                                                                         
    os.remove(f".cache-{username}")                                             
    token = util.prompt_for_user_token(username,scope,client_id=cID,client_secret=secret ,redirect_uri="http://localhost/")     
```

```python
#------------------Generate-Object---------------
ccManager = SpotifyClientCredentials(client_id=cID, client_secret=secret)
spotify = spotipy.Spotify(auth=token, client_credentials_manager=ccManager)
```

```python
#--------Try-to-get-user-data------
try:
    #---Liked-songs-----
    #topArtists = spotify.current_user_top_artists(limit=20,offset=0,time_range="long_term")
    topSongs = spotify.current_user_top_tracks(limit=50,offset=0,time_range="long_term")["items"]
    topSongs += spotify.current_user_top_tracks(limit=50,offset=50,time_range="long_term")["items"]
    #---Disliked-songs-----
    disliked = spotify.user_playlist_tracks("21fm7pkvrha54o2nt7kiyqema",playlist_id="4kIIC9gnsnRF2LDRFoIpz6",limit=100,offset=0,market=None)["items"]
    print("Success")
except:
    print("No Scope")
```

```python
#----------Format-Data-----------------LIKED SONGS-------------------
for i in range(len(topSongs)):
    print("---------------------------------")
    print("#: ", i)
    print("Song Name: ", topSongs[i]["name"])
    #print("ID: ", topSongs[i]["id"])
    print("Artist Name: ", topSongs[i]["artists"][0]["name"])
    print("Album Name: ", topSongs[i]["album"]["name"])
    print("---------------------------------")
```

```python
#----------Format-Data-----------------DISLIKED SONGS-------------------
for i in range(len(disliked)):
    print("---------------------------------")
    print("#: ", i)
    print("Song Name: ", disliked[i]["track"]["name"])
    #print("ID: ", topSongs[i]["id"])
    print("Artist Name: ", disliked[i]["track"]["artists"][0]["name"])
    print("Album Name: ", disliked[i]["track"]["album"]["name"])
    print("---------------------------------")
```

```python
#------Extract-IDS----
likedIds = []
dislikedIds = []

for i in range(len(topSongs)): 
    likedIds.append(topSongs[i]["id"]) 
    
for i in range(len(disliked)): 
    dislikedIds.append(disliked[i]["track"]["id"]) 
```

```python
#------Extract-Features----
data = spotify.audio_features(likedIds) + spotify.audio_features(dislikedIds)
for i in range(len(data)):
    del data[i]["analysis_url"]
    del data[i]["id"]
    del data[i]["track_href"]
    del data[i]["type"]
    del data[i]["uri"]
```

```python
#--Get-Feature-Names---
features = list(data[0].keys())
features.sort()
```

```python
#Rate
for i in range(len(data)):
    if i <= 99:
        data[i]["Rating"] = 1
    else:
        data[i]["Rating"] = 0
dataTB = pd.DataFrame(data)
dataTB
```

```python
#Randomize
import random
random.shuffle(data)
dataTB = pd.DataFrame(data)
dataTB
```

```python
#---Processing-Data-----

trainingData=[]
for i in range(100):
    trainingData.append([])
    for feature in features:
        trainingData[i].append(data[i].get(feature))            
            

testingData=[]
for i in range(100):
    testingData.append([])
    for feature in features:
        testingData[i].append(data[100+i].get(feature))    
```

```python
#------TARGET----
trainingTarget=[]
for i in range(100):
    trainingTarget.append(data[i].get("Rating"))            
            
testingTarget=[]
for i in range(100):
    testingTarget.append(data[100+i].get("Rating"))   

print(testingTarget)
```

```python
#print("Data is: ", data)
#print("Target outcome: ", target)
```

```python
decisionTree = tree.DecisionTreeClassifier()
decisionTree = decisionTree.fit(trainingData,trainingTarget)

dot_data = tree.export_graphviz(decisionTree, out_file=None,
                                feature_names=features,
                                filled=False,
                                special_characters=True)
graph = graphviz.Source(dot_data)
graph.render("RateMySong")
graph
```

```python
testingResult = decisionTree.predict(testingData)
print(testingResult)
```

```python
print(testingTarget)
```

```python
workingValue = 0
for i in range(len(testingResult)):
    if testingResult[i] == testingTarget[i]:workingValue += 1
print(workingValue, "/{0}".format(len(testingResult)))
```

```python
decisionTree.score(testingData,testingTarget)
```
